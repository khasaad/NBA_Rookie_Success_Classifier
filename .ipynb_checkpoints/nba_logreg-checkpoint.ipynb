## 3. Model Training

### 3.1 Extract names, labels, feature names, and values

```python
names = df['Name'].values.tolist()  # players' names
labels = df['TARGET_5Yrs'].values  # labels
features = df.drop(['TARGET_5Yrs', 'Name'], axis=1)
```

### 3.2 Normalize dataset

```python
X = MinMaxScaler().fit_transform(features)
```

### 3.3 Define classifiers and their hyperparameters

```python
classifiers = {
    'Logistic Regression': {
        'model': LogisticRegression(max_iter=1000),
        'params': {}
    },
    'Random Forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [100, 200, 300],
            'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10]
        }
    },
    'Support Vector Classifier': {
        'model': SVC(),
        'params': {
            'C': [0.1, 1, 10, 100],
            'gamma': [1, 0.1, 0.01, 0.001],
            'kernel': ['rbf', 'linear']
        }
    }
}
```

### 3.4 Training with GridSearch

```python
# Evaluate each classifier
best_classifiers = {}
for name, classifier_info in classifiers.items():
    print(f"Evaluating {name}...")
    model = classifier_info['model']
    params = classifier_info['params']
    grid_search = GridSearchCV(estimator=model, param_grid=params, cv=3, scoring='recall')
    grid_search.fit(X, labels)
    best_classifier = grid_search.best_estimator_
    best_classifiers[name] = best_classifier
    print(f"Best classifier for {name}: {best_classifier}")
    # Save the best classifier model
    model_path = os.path.join("Models", f"{"_".join(name.split())}_model.pkl")
    joblib.dump(best_classifier, model_path)
    # Create an instance of ClassifierEvaluator and call it
    evaluator = ScoreClassifier(best_classifier, X, labels)
    evaluator()
    print("_________________________________________________________________________________")
    
# Print all best classifiers' recall scores for comparison
for name, classifier in best_classifiers.items():
    print(f"Evaluating the best {name} model...")
    evaluator = ScoreClassifier(classifier, X, labels)
    evaluator()
    print("______________________________________")
```

### 3.5 Training with RandomizedSearchCV

```python
# Evaluate each classifier
best_classifiers = {}
for name, classifier_info in classifiers.items():
    print(f"Evaluating {name}...")
    model = classifier_info['model']
    params = classifier_info['params']
    grid_search = RandomizedSearchCV(estimator=model, param_distributions=params, verbose=3, n_jobs=-1, return_train_score=True, scoring="r2")
    grid_search.fit(X, labels)
    best_classifier = grid_search.best_estimator_
    best_classifiers[name] = best_classifier
    print(f"Best classifier for {name}: {best_classifier}")
    # Save the best classifier model
    model_path = os.path.join("Models", f"{"_".join(name.split())}_RSV_model.pkl")
    joblib.dump(best_classifier, model_path)
    # Create an instance of ClassifierEvaluator and call it
    evaluator = ScoreClassifier(best_classifier, X, labels)
    evaluator()

# Print all best classifiers' recall scores for comparison
for name, classifier in best_classifiers.items():
    print(f"Evaluating the best {name} model...")
    evaluator = ScoreClassifier(classifier, X, labels)
    evaluator()
```

### 5. Feature Importance

```python
# Function to get feature importances based on model type
def get_feature_importances(model):
    if isinstance(model, LogisticRegression):
        return np.abs(model.coef_[0])  # Logistic Regression coefficients
    elif isinstance(model, RandomForestClassifier):
        return model.feature_importances_  # Random Forest feature importances
    elif isinstance(model, SVC) and model.kernel == 'linear':
        return np.abs(model.coef_[0])  # SVM with linear kernel coefficients
    else:
        raise ValueError("Model type not supported for feature importances")
```

```python
def plot_feature_importances(trained_model):
    # Get feature importances based on model type
    importance = get_feature_importances(trained_model)
    
    # Normalize importances to sum up to 100%
    # importance_percent = 100.0 * (importance / np.sum(importance))
    
    # Sort feature indices by importance
    indices = np.argsort(importance)
    
    # Plotting
    plt.style.use('fivethirtyeight')
    plt.figure(figsize=(8, 6))
    plt.title(f"Feature Importance of {trained_model}")
    plt.barh(range(len(indices)), importance[indices], color='b', align='center')
    plt.yticks(range(len(indices)), [list(features.columns)[i] for i in indices])
    plt.xlabel('Importance')
    plt.tight_layout()
    plt.show()
```

#### 5.1 Feature importance of SVC with RandomizedSearchCV

```python
Support_Vector_Classifier_RSV_model = joblib.load('Models/Support_Vector_Classifier_RSV_model.pkl')
plot_feature_importances(Support_Vector_Classifier_RSV_model)
```
